```python
import pandas as pd

# load the thales stock price csv
thales = pd.read_csv('thales_stock_prices.csv')
thales.head()
```

```python
# check for nulls
thales.isnull().value_counts()
```

```python
# create boolean mask to display nulls
thalesna = thales[thales.isna().any(axis=1)]
thalesna
```
